```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month. 
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y_axis.
stdev = statistics.stdev(y_axis)
stdev
```

## Error Bars on Line Charts

### MATLAB method

```python
# Create a line chart with error bars. 
plt.errorbar(x_axis, y_axis, yerr=stdev)
plt.show()
```

```python
# Create a line chart with error bars with caps. 
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

### Object-Oriented Interface

```python
# Create a line chart with ax.plot(), with error bars. 
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Error Bars on Bar Charts

### MATLAB method

```python
# Create a bar chart with plt.bar(), with error bars.
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

### Object_Oriented Interface

```python
# Create a bar chart with ax.bar(), with error bars.
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## Changing Ticks

```python
# Create a horizonat bar chart with plt.barh(), and adjust x ticks.
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Create a horizonat bar chart with ax.barh(), and adjust x ticks.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
# ax.xaxis.set_ticks(np.arange(0, 51, step=5.0))
ax.set_xticks(np.arange(0, 51, step=5.0))
ax.invert_yaxis()
plt.show()
```

```python
# Create a horizonat bar chart with ax.barh(), and adjust x ticks.
# Set minor anumerical ticks. 
import numpy as np
from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure. 
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1. 
ax.xaxis.set_minor_locator(MultipleLocator(1))

ax.invert_yaxis()
plt.show()
```
