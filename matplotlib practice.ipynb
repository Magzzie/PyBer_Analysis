```python
# Object-Oriented Interface method. 
# # Determine the size of the chart.
# fig, ax = plt. subplots(figsize=(20, 10))
# Create the plot.
fig, ax = plt.subplots()
# Create the plot and add a label for the legend. 
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label="Boston")
# Create labels for x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
# Create a title.
ax.set_title('PyBer Fare by Month')
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Add grid lines.
ax.grid()
# Add legend. 
ax.legend()
```

## Scatter Plots

```python
# Set the x-axis to a list of strings for each month. 
x_axis = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter plot using plt.plot() with 'o' parameter.
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot using plt.scatter().
plt.scatter(x_axis, y_axis)
```

```python
# Create scatter plot and annotate.
plt.scatter(y_axis, x_axis, c='r', label='Chicago')
# Create x and y labels. 
plt.xlabel('Fare ($)')
plt.ylabel('Date')
# Create title.
plt.title('PyBer Fare by Month')
# Add a limit for the x axis data. 
plt.xlim(0, 50)
# Invert the y axis so that January is at the top. 
plt.gca().invert_yaxis()
# Add legend.
plt.legend()
```

```python
# Create a scatter chart using OOI.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

## Bubble Chart

### MATLAB method

```python
# Create bubble chart: with size of marker.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Increasing size of marker by creating a new data list of the y axis.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
# Create bubble chart
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Increasing size of marker by using Python's list comprehension technique.
# new_list =[expression for item in list if conditional]
# Create bubble chart
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

### Object-Oriented Interface

```python
# Create a bubble chart using OOI.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create a bubble chart using OOI and annotate.
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', alpha=0.8, edgecolor='black', linewidth=2, s=[i*5 for i in y_axis], label='Boston')
# Create x and y lables.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Create title.
ax.set_title('PyBer Fare by Month')
# Add x limit
ax.set_xlim(0, 50)
# Invert the y axis so that January is at the top.
ax.invert_yaxis()
# Add legend.
ax.legend()
```

## Pie Charts

### Pie charts with MATLAB method. 

```python
# Use x_axis data(months) as labels; y_axis data (total fare for each month) as values.
# Create a pie chart with MATLAB.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create a pie chart with MATLAB.
# explode values can be a tuple () or a list []
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create a pie chart with MATLAB.
# Increase the size of the pie so it's not crowded. 
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Create a pie chart with MATLAB.
# Change the color palette, use CSS colors in Matplotlib documentation.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

plt.subplots(figsize=(8, 8))
plt.pie(y_axis, 
        explode=explode_values, 
        colors=colors,
        labels=x_axis, 
        autopct='%.1f%%')

plt.show()
```

### Pie charts with Object-Oriented Interface.

```python
# Create pie chart with OOI.
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create a pie chart with OOI and annotate.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, 
       explode=explode_values, 
       shadow=True, 
       startangle=90, 
       counterclock=False,
       colors=colors, 
       labels=x_axis, 
       autopct='%.1f%%')

plt.show()
```
